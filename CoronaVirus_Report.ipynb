## Lockdown Impact 

For this section , We analyse the statewise increase/decrease in the coronavirus cases. For this section , we keep a buffer of 14 days from the day lockdown was announced. Since median number for the Coronavirus to be detected is around 5-6 days , with the maximum of 14 days. We keep a buffer of 14 days for our analysis

```
bfr_lckdwn = s1[s1['Date'] <= pd.to_datetime('8-April-2020')]
```

```
aftr_lckdwn = s1[s1['Date'] > pd.to_datetime('8-April-2020')]
```

```
cpd_bfr = bfr_lckdwn.groupby('variable')['value'].mean().to_frame()
cpd_afr = aftr_lckdwn.groupby('variable')['value'].mean().to_frame()
```

```
cpd_bfr = cpd_bfr.reset_index()
cpd_afr = cpd_afr.reset_index()
```

```
import plotly.graph_objects as go


fig = go.Figure()
fig.add_trace(go.Bar(x=cpd_bfr['variable'],
                y=cpd_bfr['value'],
                name='Before Lockdown',
                marker_color='rgb(55, 83, 109)'
                ))
fig.add_trace(go.Bar(x=cpd_afr['variable'],
                y=cpd_afr['value'],
                name='After Lockdown',
                marker_color='rgb(26, 118, 255)'
                ))

fig.update_layout(
    title='Statewise Case',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Case per Day',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=0,
        y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    barmode='group',
    bargap=0.15, # gap between bars of adjacent location coordinates.
    bargroupgap=0.1 # gap between bars of the same location coordinate.
)
fig.show()
```

```
combined_cpd = pd.merge(cpd_bfr.rename(columns = {'value':'Before'}) , cpd_afr.rename(columns = {'value':'After'}) , on = 'variable' , how = 'left')
```

```
combined_cpd['Change'] = (combined_cpd['After'] - combined_cpd['Before'])/combined_cpd['Before'] 
```

```
fig = px.bar(combined_cpd , x = 'variable' , y = 'Change')
fig.show()
```
